```python
%pylab inline
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Cities_compare = 'https://github.com/YilunCai627/comparing-baltimore-boston-household-income/raw/master/Data_Analysis.xlsx'
Diversity = pd.read_excel(Cities_compare,sheet_name = 'Cities_compare').iloc[:,6:11]
Diversity
BAL_Diversity = Diversity.iloc[0,:]
BOS_Diversity = Diversity.iloc[1,:]
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Black', 'White','Hispanic','Asian'

BAL_sizes = list(BAL_Diversity[1:5])
BOS_sizes = list(BOS_Diversity[1:5])


fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Population Diversity of Baltimore(Left) and Boston(Right)')
ax1.pie(BAL_sizes, labels=labels, autopct='%d%%', startangle=90)
ax1.axis('equal')
ax2.pie(BOS_sizes, labels=labels, autopct='%d%%', startangle=90)
ax2.axis('equal') 

plt.show()
```
